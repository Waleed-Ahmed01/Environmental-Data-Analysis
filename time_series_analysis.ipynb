## Air pollutants and Ozone trend over time:
### O3 showing largest values followed by PM 10, PM 2.5, NO2, SO2 & CO

```python
import matplotlib.pyplot as plt
import matplotlib.dates as mdates

# Plotting each pollutant's trend over time
plt.figure(figsize=(6, 4))


for col in ['PM2.5', 'PM10', 'SO2', 'NO2', 'CO', 'O3','AQI']:
    plt.plot(data_monthly.index, data_monthly[col], label=col)


plt.xlabel('Time')
plt.ylabel('Concentration (g/m or ppb)')
plt.title('Trends of Air Pollutants Over Time')

# Set the major locator to display every 1 year
plt.gca().xaxis.set_major_locator(mdates.YearLocator())
# Set the minor locator to display every month
plt.gca().xaxis.set_minor_locator(mdates.MonthLocator())
# Format the date labels for better readability
plt.gca().xaxis.set_major_formatter(mdates.DateFormatter('%Y'))

# Rotate date labels for clarity
plt.gcf().autofmt_xdate()

plt.legend()
plt.grid()
plt.tight_layout()  
plt.show()
```

## Individual time series analysis for pollutants and environmental factors to observe trends and seasonality clearly

```python
import matplotlib.pyplot as plt

# List of pollutants
variables = ['PM2.5', 'PM10', 'SO2', 'NO2', 'CO', 'O3','AQI']
# Plot each variable as a time series
fig, axes = plt.subplots(len(variables), 1, figsize=(8,10), sharex=False)

for i, var in enumerate(variables):
    axes[i].plot(data_monthly.index, data_monthly[var], label=var)
    axes[i].set_ylabel(var)
    axes[i].legend(loc='upper left')
   
axes[i].set_xlabel('Time')

plt.tight_layout()
plt.show()
```

## Wind rose plot to see the distribution of wind direction and speed
### most frequent winds in between north and north east

```python
import pandas as pd
from windrose import WindroseAxes
import matplotlib.pyplot as plt

# Wind direction and wind speed data
wind_direction = data_monthly['Wind Direction']
wind_speed = data_monthly['Wind Speed']

# Create a wind rose plot
fig = plt.figure(figsize=(5, 4))
ax = WindroseAxes.from_ax(fig=fig)

# Create the wind rose
ax.bar(wind_direction, wind_speed, bins=6, normed=True, opening=0.8, edgecolor='white')

# Add a title
ax.set_title('Wind Rose Plot', fontsize=16)

# Add a legend for wind speed bins
ax.set_legend(title="Wind Speed (m/s)", loc="best", fontsize='small')

# Show the plot
plt.show()
```

```python
import seaborn as sns
correlation_matrix=data_monthly[['Wind Direction','Wind Speed','Temperature','Humidity','Pressure','AQI','PM2.5','PM10','SO2','NO2','CO','O3']].corr()
plt.figure(figsize=(7, 7))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Matrix Between Air Quality and Environmental Factors')
plt.show()
```

## The correlation shows that:
### Pollutants are highly correlated with temperature and pressure as compared to other environmental factors
### Temperature has a negative relation to all pollutants and pressure is positively correlated.
### O3 Ozone doesnot have significant correlation with Air quality but has significant positive correlation with temperature (0.75) and negative correlation with pressure (-0.94)
### AQI (Air Quality Index) has the highest correlation with PM2.5, PM 10,CO, SO2 and then NO2
### Temperature & pressure are highly negatively correlated.
### All pollutants are highly correlated with each other

## Regression plots to analyse relation of temperature, humidity and pressure with pollutants

```python
import seaborn as sns
import matplotlib.pyplot as plt

weather_vars = ['Temperature', 'Humidity', 'Pressure']
pollutants = ['PM2.5', 'PM10', 'SO2', 'NO2', 'CO', 'O3', 'AQI']

# Scatter plots with regression lines
for pollutant in pollutants:
    plt.figure(figsize=(12, 5))
    for i, weather_var in enumerate(weather_vars, 1):
        plt.subplot(2, 3, i)
        sns.regplot(data=data_monthly, x=weather_var, y=pollutant, scatter_kws={'alpha':0.5}, line_kws={'color':'red'})
        plt.title(f'{pollutant} vs {weather_var}')
    plt.tight_layout()
    plt.show()
```
